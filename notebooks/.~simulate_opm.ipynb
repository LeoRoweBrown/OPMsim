```python
import numpy as np
import optical_systems
import dipole_source
from matplotlib import pyplot as plt 

O1 = {'NA':0.5, 'n':1, 'f': 0.180/60, 'rotation':0}  # 60x
# O1 = {'NA':0.1, 'n':1.33, 'f': 0.180/60, 'rotation':0}  # 60x

O2 = {'NA':0.95, 'n':1, 'f': 0.180/50}  # 50x
O3 = {'NA':0.6, 'n':1, 'f': 0.200/40, 'rotation':35}  # 40x
# O3 = {'NA':1, 'n':1, 'f': 0.200/40, 'rotation':0}  # 40x
TL1 = {'NA':1, 'n':1, 'f': 0.180}
TL2 = {'NA':1, 'n':1, 'f': 0.162}
# TL2 = {'NA':1, 'n':1, 'f': 0.180}

TL3 = {'NA':1, 'n':1, 'f': 0.160}
TLX = {'NA':1, 'n':1, 'f': 0.160/40}

lenses = [O1, TL1, TL2, O2, O3]
lenses = [O1, TL1, TL2, O2, O3, TL3]
# lenses = [O1, O1]
lenses = [O1]

ndipoles = 20000

source = dipole_source.DipoleSource(name='random dipoles')
source.generate_dipoles(ndipoles)#
source.classical_photoselection((0,0))
#source.add_dipoles((0,0))
#source.add_dipoles(np.pi/2,0)

## light-sheet
title = "Intensity map of 1000-dipole source with photoselection in BFP of collection objective for polarised light-sheet (out-of-plane)"

## OPM
# title = "Intensity map in BFP of collection objective for 35-degree oblique light-sheet"

na_list = [0.01, 0.1, 0.3, 0.5, 0.7, 0.95]
r = np.zeros(len(na_list))
r_t = np.zeros(len(na_list))
for n, na in enumerate(na_list):
    print("na", na)
    O1 = {'NA':na, 'n':1.0, 'f': 0.180/60, 'rotation':0}  # 60x
    lenses = [O1]
    r[n], r_t[n] = optical_systems.antisotropy_objective_system(lenses, source, title="anisotropy photoselection system", ray_count=10000)
    
plt.figure(figsize = (8,4))
plt.suptitle("Comparison of anisotropy for {0} random dipoles with photoselection (x-polarisation)".format(ndipoles))
plt.subplot(121)
plt.plot(na_list, r_t, label="Theory")
plt.plot(na_list, r)
plt.ylabel("Anisotropy (r)")
plt.xlabel("Collection NA")
plt.legend()
plt.subplot(122)
plt.plot(na_list, r_t-r, label="Theory minus simulated")
plt.ylabel("Residual anisotropy (r)")
plt.xlabel("Collection NA")
plt.tight_layout()
```
