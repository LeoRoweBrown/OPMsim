```python
from opmsim import optical_matrices
import scipy
from matplotlib import pyplot as plt

theta_1 = 70*np.pi/180

n_film_file = "C:/Users/lnr19/code/OPMsim/refractive_index_data/SiO2.txt"
n_metal_file = "C:/Users/lnr19/code/OPMsim/refractive_index_data/Ag.txt"

n_film_data = np.genfromtxt(n_film_file, delimiter='\t')
n_film_data = n_film_data[1:,:]  # remove headers 
d = 158e-9

n_metal_data = np.genfromtxt(n_metal_file, delimiter='\t')
n_metal_data = n_metal_data[1:,:]

wavelength = np.arange(300,800,0.2)*1e-9

wl_mul=1e9
n_film_wl = n_film_data[:,0]/wl_mul
n_film = n_film_data[:,1]
k_film = n_film_data[:,2]

n_metal_wl = n_metal_data[:,0]/wl_mul
n_metal = n_metal_data[:,1]
k_metal = n_metal_data[:,2]

n_film_interp = scipy.interpolate.interp1d(n_film_wl, n_film)
k_film_interp = scipy.interpolate.interp1d(n_film_wl, k_film)

n_metal_interp = scipy.interpolate.interp1d(n_metal_wl, n_metal)
k_metal_interp = scipy.interpolate.interp1d(n_metal_wl, k_metal)

# print("theta_i")

# if hasattr(wavelength, "__len__"):
#     if len(wavelength == 1):
#         wavelength = np.ones(len(theta_1))*wavelength

n_film_complex = n_film_interp(wavelength) + 1j*k_film_interp(wavelength)
n_metal_complex = n_metal_interp(wavelength) + 1j*k_metal_interp(wavelength)

r_p, r_s = optical_matrices.compute_fresnel_protected_mirror(70*np.pi/180, n_film_complex, d, n_metal_complex, wavelength)

tanpsi = np.abs(r_p)/np.abs(r_s)
psi = np.arctan(tanpsi)*180/np.pi
rho = r_p/r_s

delta = (-np.imag(np.log(rho/tanpsi)) % (2*np.pi))*180/np.pi

plt.figure()
plt.plot(wavelength, np.abs(r_p))
plt.plot(wavelength, np.abs(r_s))

plt.show()

plt.figure()
plt.plot(wavelength,psi)
plt.show()
plt.figure()
plt.plot(wavelength,delta)
plt.show()

plt.figure()
plt.plot(np.abs(n_film_complex))
plt.show()
plt.figure()
plt.plot(np.real(n_metal_complex))
plt.plot(np.imag(n_metal_complex))

plt.show()

print(n_metal_complex)
```
